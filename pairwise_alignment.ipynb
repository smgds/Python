```python
# Install Biopython: enter 'import Bio' on command line

# Import Pairwise sequence alignment module
from Bio import pairwise2

# Import format_alignment method to visualize results
from Bio.pairwise2 import format_alignment

# Sequences assigned to variables
X = "TTGACCGAGG"
Y = "TTGCCCATCG"

# globalxx method finds all the best possible alignments in the given sequences
alignments = pairwise2.align.globalxx(X, Y)

# format_alignment method to visualize results
for alignment in alignments:
    print(format_alignment(*alignment))
```
